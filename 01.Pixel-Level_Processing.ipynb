# 픽셀기반 영상처리

### 대칭

- 기하학적 측면에서 반사(reflection)
- 2차원 유클리드 공간에서의 기하학적인 변환의 하나로 $R*2$ (2차원 유클리드 공간) 위의 선형변환
- flip code
    - = 0: 상하 대칭
    -  \> 0: 좌우 대칭
    -  \< 0: 상하좌우 대칭

```python
import cv2

img_file = "Lena.png"

src = cv2.imread(img_file)
dst = cv2.flip(src, 0)

from google.colab.patches import cv2_imshow
cv2_imshow(src)
cv2_imshow(dst)
```

### 회전

- 임의의 점을 중심으로 물체를 회전

```python
import cv2

img_file = "Lena.png"
src = cv2.imread(img_file)

height, width, channel = src.shape
matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 1)

dst = cv2.warpAffine(src, matrix, (width, height))

from google.colab.patches import cv2_imshow
cv2_imshow(src)
cv2_imshow(dst)
```

### 이미지 피라미드

- 이미지의 크기를 원하는 단계까지 샘플링하는 장업 (Up/Down sampling)

```python
import cv2

img_file = "Lena.png"
src = cv2.imread(img_file)

height, width, channel = src.shape

dst_up = cv2.pyrUp(src, dstsize=(width * 2, height * 2), borderType=cv2.BORDER_DEFAULT)
dst_dn = cv2.pyrDown(src)

from google.colab.patches import cv2_imshow
cv2_imshow(dst_up)
cv2_imshow(src)
cv2_imshow(dst_dn)
```
